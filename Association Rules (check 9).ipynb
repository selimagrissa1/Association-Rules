```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules 
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    
df=pd.DataFrame(te_ary, columns=te.columns_)  
df
```

```python
sns.countplot(x='Skirt',data=df)
```

```python
sns.countplot(x='Hat',data=df)
```

```python
apriori(df, min_support=0.6)
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

```python
records = []
for i in range(0, len(df)):
    records.append([str(df.values[i,j]) for j in range(0, 20)])
    records[i] = [x for x in records[i] if x != 'nan']
```

```python
records
```

```python
te=TransactionEncoder()
te_ary=te.fit(records).transform(records)    
df2=pd.DataFrame(te_ary, columns=te.columns_)  
df2
```

```python
apriori(df2, min_support=0.1)
```

```python
frequent_itemsets=apriori(df2, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets=apriori(df2, min_support=0.1, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.1)
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```
